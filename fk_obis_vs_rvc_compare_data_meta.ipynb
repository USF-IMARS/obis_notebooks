```python
import pandas as pd
import seaborn
from tabulate import tabulate
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame , read_csv
```

```python
# variables
rvc_records_file = 'c91bedb0-5fa7-4d3d-80c4-fd4a3bbafdc0.csv'
obis_records_file = 'FKNMS.csv'
```

```python
rvc_data = pd.read_csv(rvc_records_file, low_memory=False)
obis_data = pd.read_csv(obis_records_file, low_memory=False)
```

```python
print(tabulate([
    ['# records', len(rvc_data), len(obis_data)]
], headers=['', 'RVC only', 'ALL OBIS']))
```

```python
SELECTED_CATAGORICAL_COLUMNS = ['occurrenceStatus', 'dataset_id', 'absence', 'datasetID', 'category', 'samplingEffort']
TOP_N = 3  # show just the top N
fig, axes = plt.subplots(
    nrows=len(SELECTED_CATAGORICAL_COLUMNS), ncols=2, figsize=(8,len(SELECTED_CATAGORICAL_COLUMNS)*4)
)
for plot_i, column in enumerate(SELECTED_CATAGORICAL_COLUMNS):
    print("plotting top {} values for column '{}'".format(TOP_N, column))
    axis = axes[plot_i, 0]
    counts  = rvc_data[column].value_counts()
    sns.barplot(list(counts.index)[:TOP_N], counts.values[:TOP_N], alpha=0.8, ax=axis)
    axis.title.set_text('Top {} "{}"'.format(TOP_N, column ))
    
    axis = axes[plot_i, 1]
    counts = obis_data[column].value_counts()
    sns.barplot(list(counts.index)[:TOP_N], counts.values[:TOP_N], alpha=0.8, ax=axis)
```
